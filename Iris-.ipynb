```python
#replacing categorical data with numeric values
f['species']=f['species'].replace({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
f
```

```python
#description about data
f.describe()
```

```python
#building the model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x=f.drop(columns=['species'])
y=f['species']
xt,xtest,yt,ytest=train_test_split(x,y,test_size=0.2)
model=LogisticRegression(max_iter=1000)
model.fit(xt,yt)
```

```python
#making predictions
predict=model.predict(xtest)
predict
```

```python
#finding the accuracy of the model
from sklearn.metrics import accuracy_score
accur=accuracy_score(predict,ytest)
accur
```
