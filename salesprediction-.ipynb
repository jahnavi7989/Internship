```python
import matplotlib.pyplot as plt
x=s['TV']
y=s['Sales']
plt.scatter(x,y)
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("Tv Sales")
plt.plot()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
plt.figure(figsize=(18, 6))
sns.scatterplot(x='TV',y='Radio', hue='Sales', data=s)
plt.title("TV vs Radio")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=s['Newspaper']
y=s['Sales']
plt.scatter(x,y)
plt.xlabel("Newspaper")
plt.ylabel("Sales")
plt.title("Newspaper Sales")
plt.plot()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x=s.drop(columns=['Sales'])
y=s['Sales']
xt,xtest,yt,ytest=train_test_split(x,y,test_size=0.2)
model=LinearRegression()
model.fit(xt,yt)
predicts=model.predict(xtest)
print(predicts)
```

```python
#checking accuracy of the model
from sklearn.metrics import r2_score
acc=r2_score(predicts,ytest)
acc
```
