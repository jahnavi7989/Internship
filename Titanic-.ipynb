```python
#loading dataset
import pandas as pd
s=pd.read_csv('tested.csv')
s
```

```python
#replacing categorical data with numeric values
s['Sex']=s['Sex'].replace({'male':0,'female':1})
s
```

```python
#removing duplicates
s=s.drop_duplicates()
s['Embarked']=s['Embarked'].replace({'Q':0,'S':1,'C':2})
s
```

```python
#filling nan values with mean values
s['Age']=s['Age'].fillna(s['Age'].mean())
s
```

```python
#conveting type of name to str
s['Name']=s['Name'].astype('str')
s.dtypes
```

```python
#converting categorical data to numeric values
def title(name):
    if 'Mrs' in name:
        return 0
    elif 'Miss' in name:
        return 1
    else:
        return 2
s['Name']=s['Name'].apply(title)  
s
```

```python
#filling null values of fare with mean value
s['Fare']=s['Fare'].fillna(s['Fare'].mean())
s
```

```python
#visualise data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
s=pd.read_csv('tested.csv')
sns.set(style="whitegrid")
sns.countplot(x='Survived', data=s, palette='Set1')
plt.title('Passenger Survival Distribution (0: Not Survived, 1: Survived)')
plt.xlabel('Survival')
plt.ylabel('no of passengers')
plt.show()
```

```python
#dropping unnecessary data 
x=s.drop(columns=['Survived','Ticket','Cabin'])
x
```

```python
#output data
y=s['Survived']
y
```

```python
#building model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

xt,xtest,yt,ytest=train_test_split(x,y,test_size=0.2)
model=LogisticRegression(max_iter=1000)
model.fit(xt,yt)
```

```python
#predicting values
predicts=model.predict(xtest)
predicts
```

```python
#checking accuracy of the model
from sklearn.metrics import accuracy_score
acc=accuracy_score(predicts,ytest)
acc
```
